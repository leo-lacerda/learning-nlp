```python
from transformers import pipeline
```

```python
question_answerer = pipeline("question-answering")
question_answerer(
    question = "Where do I work?",
    context = "My name is Leo Lacerda and I'm a Professor at University of Fortaleza, in Fortaleza, Ceara, Brazil",
)
```
